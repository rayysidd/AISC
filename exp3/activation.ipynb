```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1=0.1
x2=-0.2
x3=-0.3

w1=0.2
w2=0.3
w3=0.6
```

```python
bias = -0.3
```

```python
z = (x1*w1) + (x2*w2) + (x3*w3) + bias
z
```

```python
# Activation functions
def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.1):
    return np.where(x >= 0, x, alpha * x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
y_tanh,y_relu,y_leaky,y_sigmoid = tanh(z),relu(z),leaky_relu(z),sigmoid(z)
```

```python
y = [y_tanh,y_relu,y_leaky,y_sigmoid]
```

```python
print(y)
```

```python
# Input range for plotting
z_range = np.linspace(-5, 5, 500)
z_point = (x1*w1) + (x2*w2) + (x3*w3) + bias

# Compute outputs
y_tanh = tanh(z_range)
y_relu = relu(z_range)
y_leaky = leaky_relu(z_range)
y_sigmoid = sigmoid(z_range)

# Compute single point values
y_tanh_p = tanh(z_point)
y_relu_p = relu(z_point)
y_leaky_p = leaky_relu(z_point)
y_sigmoid_p = sigmoid(z_point)
```

```python
# Plot again with annotations for the activation outputs at z = -0.52

plt.figure(figsize=(8,6))
plt.plot(z_range, y_tanh, label="tanh", linewidth=2)
plt.plot(z_range, y_relu, label="ReLU", linewidth=2)
plt.plot(z_range, y_leaky, label="Leaky ReLU (=0.01)", linewidth=2)
plt.plot(z_range, y_sigmoid, label="sigmoid", linewidth=2)

# Mark and annotate each activation output
plt.scatter([z_point], [y_tanh_p], color="blue", s=80, zorder=5)
plt.annotate(f"{y_tanh_p:.3f}", (z_point, y_tanh_p), textcoords="offset points", xytext=(-10,10), color="blue")

plt.scatter([z_point], [y_relu_p], color="orange", s=80, zorder=5)
plt.annotate(f"{y_relu_p:.3f}", (z_point, y_relu_p), textcoords="offset points", xytext=(-10,10), color="orange")

plt.scatter([z_point], [y_leaky_p], color="green", s=80, zorder=5)
plt.annotate(f"{y_leaky_p:.3f}", (z_point, y_leaky_p), textcoords="offset points", xytext=(-10,10), color="green")

plt.scatter([z_point], [y_sigmoid_p], color="red", s=80, zorder=5)
plt.annotate(f"{y_sigmoid_p:.3f}", (z_point, y_sigmoid_p), textcoords="offset points", xytext=(-10,10), color="red")

plt.title("Activation Functions with z = -0.52", fontsize=14)
plt.xlabel("z")
plt.ylabel("y")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```
